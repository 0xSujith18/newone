```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("student_performance.csv")
```

```python
print("=" * 60)
print("        STUDENT PERFORMANCE ANALYZER REPORT")
print("=" * 60)
```

```python
print("\n DATASET OVERVIEW")
print("-" * 40)

total_students = len(df)

total_columns = len(df.columns)

unique_departments = df['department'].nunique()

print("Total Number of Students :", total_students)
print("Total Number of Columns  :", total_columns)
print("Unique Departments       :", unique_departments)
```

```python
print("\n DEPARTMENT WISE ANALYSIS")
print("-" * 40)

dept_count = df['department'].value_counts()

print(dept_count)
```

```python
print("\n SCORE ANALYSIS")
print("-" * 40)

highest_math = df['math_score'].max()
lowest_math = df['math_score'].min()
average_math = df['math_score'].mean().round(2)

average_programming = df['programming_score'].mean().round(2)

average_attendance = df['attendance_percentage'].mean().round(2)

print("Highest Math Score           :", highest_math)
print("Lowest Math Score            :", lowest_math)
print("Average Math Score           :", average_math)
print("Average Programming Score    :", average_programming)
print("Average Attendance Percentage:", average_attendance)
```

```python
print("\n GENDER WISE TOTAL SCORE ANALYSIS")
print("-" * 40)

df['total_score'] = df['math_score'] + df['programming_score']
gender_avg = df.groupby('gender')['total_score'].mean().round(2)

print(gender_avg)
```

```python
print("\n TOP 5 PERFORMERS")
print("-" * 40)

top_students = df.nlargest(5, 'total_score')

print(top_students[['name', 'department', 'total_score']])
```

```python
print("\n ATTENDANCE ANALYSIS")
print("-" * 40)

attendance_data = df[['name', 'department', 'attendance_percentage']]

print(attendance_data)

# Check attendance condition
if (df['attendance_percentage'] > 75).all():
    print("\n All students have attendance above 75%")
else:
    print("\n Some students have attendance below 75%")
```

```python
print("\n" + "=" * 60)
print("          REPORT GENERATED SUCCESSFULLY")
print("=" * 60)
```
