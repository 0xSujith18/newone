```python
# 1
import pandas as pd

# ETL stands for:
# Extract – Collect data from source systems.

# Example: Read today's sales data from a CSV file.
df = pd.read_csv("messy_sales_data.csv")

# Transform – Clean and modify the data.

# Example:
# Remove missing values.
# Remove duplicate records.
# Convert date columns to proper format.
df = df.dropna()
df = df.drop_duplicates()

# Load – Store the processed data into a database or data warehouse.

df.to_csv("clean_sales.csv", index=False)
```

```python
# 2
# dropna() removes rows containing missing values.
# Original rows = 500
# Remaining rows = 412

# Rows removed:

# 500 - 412 = 88

# Conclusion: 88 rows contained at least one missing value and were removed.
```

```python
# 3
df = df.drop_duplicates(
    subset=['customer_name', 'product']
)
```

```python
# 1. fillna(0)

# Replaces all missing values with 0.

df['col'] = df['col'].fillna(0)

# Use when:

# Missing value logically means zero.
# Example: Number of items sold.
# 2. fillna(df['col'].median())

# Replaces missing values with the column median.

df['col'] = df['col'].fillna(df['col'].median())

# Use when:

# Column contains numerical data.
# Want to avoid changing data distribution.
# Useful when outliers exist.

# Missing value becomes 20 instead of 0.
```

```python
import requests

url = "https://api.openweathermap.org/data/2.5/weather"
params = {
    "q": "Delhi",
    "appid": "YOUR_API_KEY",
    "units": "metric"
}
response = requests.get(url, params=params)

data = response.json()

if response.status_code == 200:
    print("Temperature:", data["main"]["temp"], "°C")
else:
    print(f"Error: {data.get('message', 'Unknown error')}")
    print("Please ensure you have a valid API key and that the API call is correct.")
```
