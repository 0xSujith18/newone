```python
import pandas as pd
import re

invoices = [
    "Invoice#1001 Customer: John Amount Rs.2500 Date 01/05/2026",
    "Inv 1002 - Mary paid 3200 on 02-05-2026",
    "Invoice No:1003 Customer=David Total=4500 Date=03/05/2026",
    "1004 Sarah Rs 2800 04/05/2026",
    "Invoice1005 Customer Alex Amount 3900 Date 05/05/2026"
]

def parse_invoice(text):
    invoice_id = re.search(r'100\d+', text)
    numbers = re.findall(r'\d+', text)
    amount = None
    for num in numbers:
        if len(num) == 4 and not num.startswith("100"):
            amount = num
            break
    return {
        "Invoice_ID": invoice_id.group() if invoice_id else None,
        "Raw_Text": text,
        "Amount": amount
    }

cleaned_data = [parse_invoice(inv) for inv in invoices]
df = pd.DataFrame(cleaned_data)
print(df)
df["Amount"] = pd.to_numeric(df["Amount"])

print("\nTotal Amount =", df["Amount"].sum())
print("Average Amount =", df["Amount"].mean())
```
